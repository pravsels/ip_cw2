```python
def poisson_matrix(n, m):
    
    diag = scipy.sparse.lil_matrix((m, m))
    diag.setdiag(-1, -1)
    diag.setdiag(4)
    diag.setdiag(-1, 1)
    
    A = scipy.sparse.block_diag([diag] * n).tolil()
    
    A.setdiag(-1, 1*m)
    A.setdiag(-1, -1*m)
    
    return A
```

```python
def zero_matrix(n, m):
    
    diag = scipy.sparse.lil_matrix((m, m))
    diag.setdiag(0, 0)
    diag.setdiag(0)
    diag.setdiag(0, 0)
    
    A = scipy.sparse.block_diag([diag] * n).tolil()
    
    A.setdiag(0, 1*m)
    A.setdiag(0, -1*m)
    
    return A
```

```python
z = zero_matrix(3, 3)
z.todense()
```

```python
def membrane_interpolation(src, target, mask):
    
    y_max, x_max = target.shape
    y_min, x_min = 0, 0
    
    x_range = x_max - x_min
    y_range = y_max - y_min 
    
    plt.imshow(src, cmap='gray')
    plt.show()
    
    mask = mask[y_min:y_max, x_min:x_max]
    mask[mask != 0] = 1
    
    plt.imshow(mask, cmap='gray')
    plt.show()

    
    A = poisson_matrix(y_range, x_range)
    
    B = zero_matrix(y_range, x_range)
    B = B.tocsc()
    
    # set the region outside the mask to identity 
    for y in range(1, y_range - 1):
        for x in range(1, x_range - 1):
            if mask[y, x] == 0:
                k = x + y * x_range
                
                A[k, k] = 1
                A[k, k+1] = 0
                A[k, k-1] = 0
                
                A[k, k+x_range] = 0
                A[k, k-x_range] = 0
                
    
    A = A.tocsc()
    
    flat_mask = mask.flatten()
    
        
    flat_src = src[y_min:y_max, x_min:x_max].flatten()
    target_flat = target[y_min:y_max, x_min:x_max].flatten()
    
    B = B.dot(flat_src)

    # outside the mask
    # f = t
    print('shape of B : ', B.shape)
    print('shape of flat mask : ', flat_mask.shape)
    print('shape of flat target : ', target_flat.shape)
    B[flat_mask==0] = target_flat[flat_mask==0]

    X = spsolve(A, B)
    X = X.reshape((y_range, x_range))

    X[X > 255] = 255
    X[X < 0] = 0

    X = X.astype('uint8')

    target[y_min:y_max, x_min:x_max] = X
        
        
    return target
```

```python
def poisson_matrix(n, m):
    
    diag = scipy.sparse.lil_matrix((m, m))
    diag.setdiag(-1, -1)
    diag.setdiag(4)
    diag.setdiag(-1, 1)
    
    A = scipy.sparse.block_diag([diag]*n).tolil()
    
    A.setdiag(-1, 1*m)
    A.setdiag(-1, -1*m)
    
    return A
```

```python
poisson_matrix(3, 3).todense()
```

```python
def poisson_blend(src, target, mask, offset):
    
    y_max, x_max = target.shape
    y_min, x_min = 0, 0
    
    x_range = x_max - x_min
    y_range = y_max - y_min 
    
    M = np.float32([[1, 0, offset[0]], [0, 1, offset[1]]])
    src = cv.warpAffine(src, M, (x_range, y_range))
    
    plt.imshow(src, cmap='gray')
    plt.show()
    
    mask = mask[y_min:y_max, x_min:x_max]
    mask[mask != 0] = 1
    
    plt.imshow(mask, cmap='gray')
    plt.show()

    
    A = poisson_matrix(y_range, x_range)
    
    # for dela g
    laplacian = A.tocsc()
    
    # only apply laplacian to mask region, set outside region to identity
    for y in range(1, y_range - 1):
        for x in range(1, x_range - 1):
            if mask[y, x] == 0:
                k = x + y * x_range
                
                A[k, k] = 1
                A[k, k+1] = 0
                A[k, k-1] = 0
                A[k, k+x_range] = 0
                A[k, k-x_range] = 0
                
    
    A = A.tocsc()
    
    flat_mask = mask.flatten()
    
        
    flat_src = src[y_min:y_max, x_min:x_max].flatten()
    target_flat = target[y_min:y_max, x_min:x_max].flatten()

    # div(v) = delta(g)
    alpha = 1.
    B = laplacian.dot(flat_src)*alpha

    # outside the mask
    # f = t
    B[flat_mask==0] = target_flat[flat_mask==0]

    X = spsolve(A, B)
    X = X.reshape((y_range, x_range))

    X[X > 255] = 255
    X[X < 0] = 0

    X = X.astype('uint8')

    target_cp = np.copy(target)
    target_cp[y_min:y_max, x_min:x_max] = X
        
        
    return target_cp
```

```python
offset = offset_x, offset_y
print(offset)
pb_result = poisson_blend(src_image, target_image, target_mask_image, offset)
```

```python
plt.imshow(pb_result, cmap='gray')
plt.show()
```

```python
cv.imwrite('./seamless_cloning.jpg', pb_result)
```

```python
def poisson_blend_with_mixed_gradients(src, target, mask, offset):
    
    y_max, x_max = target.shape
    y_min, x_min = 0, 0
    
    x_range = x_max - x_min
    y_range = y_max - y_min 
    
    M = np.float32([[1, 0, offset[0]], [0, 1, offset[1]]])
    src = cv.warpAffine(src, M, (x_range, y_range))
    
    plt.imshow(src, cmap='gray')
    plt.show()
    
    plt.imshow(target, cmap='gray')
    plt.show()
    
    mask = mask[y_min:y_max, x_min:x_max]
    mask[mask != 0] = 1
    
    plt.imshow(mask, cmap='gray')
    plt.show()

    
    A = poisson_matrix(y_range, x_range)
    
    # for dela g
    laplacian = A.tocsc()
    
    # only apply laplacian to mask region, set outside region to identity 
    for y in range(1, y_range - 1):
        for x in range(1, x_range - 1):
            if mask[y, x] == 0:
                k = x + y * x_range
                
                A[k, k] = 1
                A[k, k+1] = 0
                A[k, k-1] = 0
                A[k, k+x_range] = 0
                A[k, k-x_range] = 0
                
    
    A = A.tocsc()
    
    flat_mask = mask.flatten()
    
        
    flat_src = src[y_min:y_max, x_min:x_max].flatten()
    target_flat = target[y_min:y_max, x_min:x_max].flatten()

    # div(v) = greater of abs of delta(g) or delta(f*)
    alpha = 1.
    g = laplacian * flat_src
    f_star = laplacian * target_flat

    B = np.where(np.abs(f_star) > np.abs(g), f_star, g)

    # outside the mask
    # f = t
    B[flat_mask==0] = target_flat[flat_mask==0]

    X = spsolve(A, B)
    X = X.reshape((y_range, x_range))
    
    X[X > 255] = 255
    X[X < 0] = 0

    X = X.astype('uint8')

    target_cp = np.copy(target)
    target_cp[y_min:y_max, x_min:x_max] = X
        
        
    return target_cp
```

```python
offset = offset_x, offset_y
pb_result = poisson_blend_with_mixed_gradients(src_image, target_image, target_mask_image, offset)
```

```python
plt.imshow(pb_result, cmap='gray')
plt.show()
```

```python
cv.imwrite('./mixed_gradients.jpg', pb_result)
```
